```python
import pandas as pd
import plotly.graph_objects as go
from pycoingecko import CoinGeckoAPI
```

```python
cg = CoinGeckoAPI()
```

```python
data = cg.get_coin_market_chart_by_id(id="bitcoin", vs_currency="USD", days=30)
```

```python
data
```

```python
bitcoin_data = pd.DataFrame(data['prices'], columns= ["Timestamp", "Price"])
```

```python
bitcoin_data
```

```python
bitcoin_data["Date"] = pd.to_datetime(bitcoin_data['Timestamp'], unit='ms')
```

```python
bitcoin_data
```

```python
candlestick_data = bitcoin_data.groupby(bitcoin_data.Date.dt.date).agg({"Price":['max', 'min', 'first', 'last']})
```

```python
candlestick_data
```

```python
fig = go.Figure(data=[go.Candlestick(x=candlestick_data.index,
                                    open=candlestick_data["Price"]['first'],
                                    close=candlestick_data["Price"]["last"],
                                    high=candlestick_data["Price"]["max"],
                                    low=candlestick_data["Price"]["min"])
                     ])
```

```python
fig.update_layout(xaxis_title="Date", yaxis_title="Price", title="Candlestick Graph of the market price of Bitcoin in the last 30 days in USD")
```
